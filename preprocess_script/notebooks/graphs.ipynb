```python
# Data for first bar chart (Egret Model comparison with and without temperature)
metrics_egret = ['Loss', 'MSE', 'MAE', 'R²']
egret_with_temp = [0.3, 0.033, 0.023, 0.67]
egret_without_temp = [0.8, 0.19, 0.15, 0.11]

# Create a bar chart for Egret model (with and without temperature)
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

x_egret = np.arange(len(metrics_egret))  # the label locations
width = 0.35  # the width of the bars

ax.bar(x_egret - width/2, egret_without_temp, width, label='Without Temp')
ax.bar(x_egret + width/2, egret_with_temp, width, label='With Temp')

# Set properties for the Egret comparison chart
ax.set_title('Egret Model - With vs Without Temperature')
ax.set_xticks(x_egret)
ax.set_xticklabels(metrics_egret)
ax.legend(loc='best')
ax.set_ylabel('Scores')

plt.tight_layout()
plt.show()

# Data for second bar chart (Linear and Random Forest Models)
metrics_rf_linear = ['MSE', 'MAE', 'R²']
linear_temp_only = [0.19, 0.15, 0.11]
linear_reaction_only = [0.19, 0.15, 0.11]
linear_temp_reaction = [0.033, 0.023, 0.67]

rf_temp_only = [0.0181, 0.1058, 0.0081]
rf_reaction_only = [0.0098, 0.0605, 0.4624]
rf_temp_reaction = [0.0097, 0.0606, 0.4703]

# Create a 3x1 bar chart for Temp Only, Reaction Only, and Temp + Reaction comparison
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

x_rf = np.arange(len(metrics_rf_linear))  # the label locations

# First chart: Temp Only comparison
ax[0].bar(x_rf - width/2, linear_temp_only, width, label='Linear Regression')
ax[0].bar(x_rf + width/2, rf_temp_only, width, label='Random Forest')
ax[0].set_title('Temperature Only Comparison')
ax[0].set_xticks(x_rf)
ax[0].set_xticklabels(metrics_rf_linear)
ax[0].legend(loc='best')

# Second chart: Reaction Only comparison
ax[1].bar(x_rf - width/2, linear_reaction_only, width, label='Linear Regression')
ax[1].bar(x_rf + width/2, rf_reaction_only, width, label='Random Forest')
ax[1].set_title('Reaction Only Comparison')
ax[1].set_xticks(x_rf)
ax[1].set_xticklabels(metrics_rf_linear)
ax[1].legend(loc='best')

# Third chart: Temperature and Reaction comparison
ax[2].bar(x_rf - width/2, linear_temp_reaction, width, label='Linear Regression')
ax[2].bar(x_rf + width/2, rf_temp_reaction, width, label='Random Forest')
ax[2].set_title('Temperature and Reaction Comparison')
ax[2].set_xticks(x_rf)
ax[2].set_xticklabels(metrics_rf_linear)
ax[2].legend(loc='best')

# Set common properties
for i in range(3):
    ax[i].set_ylabel('Scores')

plt.tight_layout()
plt.show()
```
