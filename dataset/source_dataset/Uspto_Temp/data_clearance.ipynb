```python
from rdkit import Chem
from rdkit.Chem import Draw

# Splitting the SMILES string into reactant and product
smiles_data = "CC(C)(C)OC(=O)n1c(-c2cnc(Nc3cccc(O)c3)o2)cc2ccccc21>>Oc1cccc(Nc2ncc(-c3cc4ccccc4[nH]3)o2)c1"
reactant_smiles, product_smiles = smiles_data.split(">>")

# Generating molecule objects from SMILES
reactant_mol = Chem.MolFromSmiles(reactant_smiles)
product_mol = Chem.MolFromSmiles(product_smiles)

# Drawing the molecules
reactant_img = Draw.MolToImage(reactant_mol, size=(300, 300))
product_img = Draw.MolToImage(product_mol, size=(300, 300))

reactant_img, product_img
```
